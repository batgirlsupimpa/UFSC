```python
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima_process import ArmaProcess
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

def plot_arma_behavior(p_coeffs, q_coeffs, title):
    # Criando o processo ARMA
    # Nota: No statsmodels, os coeficientes AR devem ter sinal invertido e incluir o 1 do lag 0
    ar = np.r_[1, -np.array(p_coeffs)]
    ma = np.r_[1, np.array(q_coeffs)]
    processo = ArmaProcess(ar, ma)

    # Gerando os dados
    y = processo.generate_sample(nsample=1000)

    fig, axes = plt.subplots(1, 2, figsize=(15, 4))
    fig.suptitle(title, fontsize=16)

    # Plot FAC (ACF)
    plot_acf(y, ax=axes[0], lags=20, title="Função de Autocorrelação (FAC)")
    # Plot FACP (PACF)
    plot_pacf(y, ax=axes[1], lags=20, title="Função de Autocorrelação Parcial (FACP)")
    plt.show()

# --- Executando para os modelos da questão ---

# a) ARMA(3, 2) - Exemplo de coeficientes estáveis
plot_arma_behavior([0.5, -0.2, 0.1], [0.4, 0.2], "Modelo a) ARMA(3, 2)")

# b) ARMA(0, 3) - Este é um modelo MA(3) puro
plot_arma_behavior([], [0.7, 0.4, 0.1], "Modelo b) ARMA(0, 3) ou MA(3)")

# c) ARMA(4, 7)
plot_arma_behavior([0.4, -0.1, 0.2, -0.1], [0.3, 0.2, 0.1, 0.1, 0.05, 0.02, 0.01], "Modelo c) ARMA(4, 7)")
```
